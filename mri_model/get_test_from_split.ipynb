```python


file = "/Users/emma/Desktop/QUEENS/THESIS/contrastive/mri_model/splits.json"
import json
with open(file) as f:
    splits = json.load(f)
print(len(splits[0]["train"]))
print(len(splits[0]["val"]))
print(len(splits))
print(splits[0].keys())

ids = set()
for i in range(len(splits)):
    for id in splits[i]["train"]:
        ids.add(id)
    for id in splits[i]["val"]:
        ids.add(id)

print(len(ids))
# print(ids)
```

```python
import csv
test = set()
with open("/Users/emma/Desktop/QUEENS/THESIS/contrastive/mri_data/picai_labels/clinical_information/marksheet.csv") as f:
    reader = csv.DictReader(f)
    rows = list(reader)
    for row in rows:
        pid_study = f"{row['patient_id']}_{row['study_id']}"
        if not pid_study in ids:
            test.add(pid_study)
print(len(test))
```

```python
with open("picai_test.json", "w") as json_file:
    json.dump(sorted(test), json_file, indent=2)
```
